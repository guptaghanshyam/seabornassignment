1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Ans:Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn along with their uses:

Scatter Plot:
Use: It represents the relationship between two continuous variables. Each point on the plot represents an observation in the dataset.
Line Plot:
Use: It shows the relationship between two continuous variables over a continuous interval. It is commonly used to visualize trends and patterns over time.
Bar Plot:
Use: It displays categorical data using rectangular bars. It can be used to compare different categories or show the distribution of a single category.
Histogram:
Use: It provides a visual representation of the distribution of a continuous variable. It divides the data into bins and shows the frequency of observations within each bin.
Heatmap:
Use: It is a 2D representation of data where the values are represented as colors. It is commonly used to visualize correlations between variables in a dataset.
These are just a few examples of the plots available in Seaborn. The library offers a wide range of other plots, such as box plots, violin plots, swarm plots, and more, which can be used to visualize different types of data and relationships.

2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri_data = sns.load_dataset("fmri")

sns.lineplot(data=fmri_data ,x="timepoint" ,y="signal",hue="event" ,style ="region")

import matplotlib.pyplot as plt

plt.show()
```

3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.boxplot(data=titanic ,x = 'pclass' ,y='age' )
plt.title("Box Plot of Age based on Pclass")
plt.show()

sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot of Fare based on Pclass")
plt.show()
```

4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
diamonds =sns.load_dataset("diamonds")

sns.histplot(data=diamonds , x='price' , hue = 'cut')

import matplotlib.pyplot as plt
plt.show()
```

5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns 

iris = sns.load_dataset("iris")

sns.pairplot(data = iris ,hue='species')

import matplotlib.pyplot as plt

plt.show()
```

Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

flights= sns.load_dataset("flights")

flights_pivot = flights.pivot("month", "year", "passengers")

sns.heatmap(data=flights_pivot)

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap")

import matplotlib.pyplot as plt
plt.show()
```
